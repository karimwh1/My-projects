```python
import json
import requests
from pprint import pprint
import csv
import pandas as pd
```

```python
url = 'https://www.vmh.life//_api/biomarkers/'
number_pages = 1
df = pd.DataFrame(columns=['name_ofd','type_ofd','notes_ofd','phenotype','organ_affected','cheBlId_ofb','description_ofb','fullName_ofb','inchiKey_ofb','inchiString_ofb','keggId_ofb','casRegistry_ofb','value_ofb','normal_conc_ofb','ramedis_ofb','range_conc_ofb'])
```

```python
for i in range(1, number_pages+3):
    response = requests.request("GET", f'https://www.vmh.life//_api/biomarkers/?page={i}')
    response = response.json()
    for bio in response['results']:
        name_ofd = bio['iem']['name']
        type_ofd = bio['iem']['dtype']
        notes_ofd = bio['iem']['notes']
        phenotype = bio['iem']['phenotype']
        organ_affected = bio['iem']['organ']
        cheBlId_ofb = bio['metabolite']['cheBlId']
        description_ofb = bio['metabolite']['description']
        fullName_ofb = bio['metabolite']['fullName']
        inchiKey_ofb = bio['metabolite']['inchiKey']
        inchiString_ofb = bio['metabolite']['inchiString']
        keggId_ofb = bio['metabolite']['keggId']
        casRegistry_ofb = bio['metabolite']['casRegistry']
        value_ofb = bio['value']
        normal_conc_ofb = bio['normalConcentration']
        ramedis_ofb = bio['ramedis']
        range_conc_ofb = bio['rangeConcentration']
        df = df.append({'name_ofd':name_ofd,'type_ofd':type_ofd,'notes_ofd':notes_ofd,'phenotype':phenotype,'organ_affected':organ_affected,'cheBlId_ofb':cheBlId_ofm,'description_ofb':description_ofb,'fullName_ofb':fullName_ofb,'inchiKey_ofb':inchiKey_ofb,'inchiString_ofb':inchiString_ofb,'keggId_ofb':keggId_ofb,'casRegistry_ofb':casRegistry_ofb,'value_ofb':value_ofb,'normal_conc_ofb':normal_conc_ofb,'ramedis_ofb':ramedis_ofb,'range_conc_ofb':range_conc_ofb},ignore_index= True)
```
