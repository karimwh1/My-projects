```python
import json
import requests
from pprint import pprint
import csv
import pandas as pd
```

```python
url = 'https://www.vmh.life//_api/biomarkers/'
number_pages = 1
df = pd.DataFrame(columns=['name_of_disease','abbreviation_of_disease','type_of_disease','notes_of_disease','phenotype','organ_affected','cheblid_of_biomarker','description_of_biomarker','fullname_of_biomarker','abbreviation_of_biomarker','inchikey_of_biomarker','inchistring_of_biomarker','keggid_of_biomarker','casregistry_of_biomarker','value_of_biomarker','normal_conc_of_biomarker','ramedis_of_biomarker','range_conc_of_biomarker','biocyc_biomarker','smile_biomarker','hmdb_biomarker','reconMap3_biomarker','food_db_biomarker','pubChemId_biomarker'])
```

```python
responce = requests.get(url).json()
pprint(responce['results'][1])
```

 pubChemId, smile, hmdb, reconMap3, food_db

```python
for i in range(1, number_pages+3):
    response = requests.request("GET", f'https://www.vmh.life//_api/biomarkers/?page={i}')
    response = response.json()
    for bio in response['results']:
        abbreviation_of_disease = bio['iem']['abbreviation']
        name_of_disease = bio['iem']['name']
        type_of_disease = bio['iem']['dtype']
        notes_of_disease = bio['iem']['notes']
        phenotype = bio['iem']['phenotype']
        organ_affected = bio['iem']['organ']
        cheblid_of_biomarker = bio['metabolite']['cheBlId']
        description_of_biomarker = bio['metabolite']['description']
        fullname_of_biomarker = bio['metabolite']['fullName']
        abbreviation_of_biomarker = bio['metabolite']['abbreviation']
        inchikey_of_biomarker = bio['metabolite']['inchiKey']
        inchistring_of_biomarker = bio['metabolite']['inchiString']
        keggid_of_biomarker = bio['metabolite']['keggId']
        casregistry_of_biomarker = bio['metabolite']['casRegistry']
        value_of_biomarker = bio['value']
        normal_conc_of_biomarker = bio['normalConcentration']
        ramedis_of_biomarker = bio['ramedis']
        range_conc_of_biomarker = bio['rangeConcentration']
        biocyc_biomarker = bio['metabolite']['biocyc']
        smile_biomarker = bio['metabolite']['smile']
        hmdb_biomarker = bio['metabolite']['hmdb']
        reconMap3_biomarker = bio['metabolite']['reconMap3']
        food_db_biomarker = bio['metabolite']['food_db']
        pubChemId_biomarker = bio['metabolite']['pubChemId']
        df = df.append({'abbreviation_of_disease':abbreviation_of_disease,'name_of_disease':name_of_disease,'type_of_disease':type_of_disease,'notes_of_disease':notes_of_disease,'phenotype':phenotype,'organ_affected':organ_affected,'cheblid_of_biomarker':cheblid_of_biomarker,'description_of_biomarker':description_of_biomarker,'fullname_of_biomarker':fullname_of_biomarker,'inchikey_of_biomarker':inchikey_of_biomarker,'inchistring_of_biomarker':inchistring_of_biomarker,'keggid_of_biomarker':keggid_of_biomarker,'casregistry_of_biomarker':casregistry_of_biomarker,'value_of_biomarker':value_of_biomarker,'normal_conc_of_biomarker':normal_conc_of_biomarker,'ramedis_of_biomarker':ramedis_of_biomarker,'range_conc_of_biomarker':range_conc_of_biomarker,'biocyc_biomarker':biocyc_biomarker,'pubChemId_biomarker':pubChemId_biomarker,'smile_biomarker':smile_biomarker,'hmdb_biomarker':hmdb_biomarker,'reconMap3_biomarker':reconMap3_biomarker,'food_db_biomarker':food_db_biomarker,'abbreviation_of_biomarker':abbreviation_of_biomarker},ignore_index= True)
```
