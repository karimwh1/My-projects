# Example: Using the VMH API to extract all reactions contained in the database

```python
# Import statements
import requests
import pandas as pd
import math
import io
```

```python
# URL of specific API we want to access
url = "https://www.vmh.life/_api/reactions/"
```

```python
# The information is saved on multiple pages, so in order to get all the information we need to query all available pages and extract the data
page_1 = url + "?page=1"
page_1
```

```python
# Here we perform the query to access the first page and have a look at the data
response = requests.request("GET", page_1)
response = response.json()
```

```python
response
```

```python
# The response is a JSON file which we can use kind of like a dictionary after loading it in python. 
# To see which parts the response consists of, we can look at the keys:
response.keys()
```

```python
# The response is a JSON file which we can use kind of like a dictionary after loading it in python. 
# To see which parts the response consists of, we can look at the keys:
response.keys()
```

```python
# With pandas we can read in the results part as a dataframe so we can continue working with it in an easier way
reactions = pd.DataFrame(response["results"])
reactions
```

```python
# We can see that a total of 50 reactions are saved on one page
len(reactions)
```

```python
# In total there are 19313 reactions contained within VMH
response['count']
```

```python
# Now that we know how many results are on one page and how many results there are in total, we can calculate how many pages we have to query in total
number_pages = math.ceil(response["count"] / len(response["results"]))
number_pages
```

```python
# We can access the url for the next page via the component 'next' in the response
next_page = response["next"]
next_page
```

```python
# Now we can loop through all pages to query them
for i in range(1, number_pages+1):
    response = requests.request("GET", f'https://www.vmh.life/_api/reactions/?page={i}')
    response = response.json()

    if i == 1:
        reactions = pd.DataFrame(response["results"])
    else:
        reactions_to_add = pd.DataFrame(response["results"])
        reactions = pd.concat([reactions, reactions_to_add])

    # For the purpose of this example this loop will stop after 10 pages, as it takes a long time to iterate through all 387 pages
    if i == 10:
        break
```

```python
# Let's have a look at the resulting dataframe
reactions
```

```python
# We have to reset the index as it does not count continuously, but in iterations of 50
reactions.reset_index(inplace=True, drop=True)
reactions
```

```python
# We can get some more information about the dataframe, e.g. how many rows it has and how the columns are named
reactions.info()
```

```python
# As this might be useful for later we want to save this information in a txt file
buffer = io.StringIO()
reactions.info(buf=buffer)
reactions_info = buffer.getvalue()

with open('vmh_example_extraction_reactions_info.txt', 'w') as file:
    file.write(reactions_info)
```

```python
# And finally we want to save the extracted data in our dataframe as a csv file
reactions.to_csv('vmh_example_extraction_reactions_all.csv')
```
